# Seaborn
Seaborn is a Python library built on top of Matplotlib.
It makes beautiful and informative statistical charts
with very less code.

- Cleaner and better looking charts than Matplotlib
- Built in themes and color palettes
- Works directly with Pandas DataFrames
- Great for statistical visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
print("Seaborn version:", sns.__version__)
```

## Dataset
Same student dataset used throughout this notebook.

```python
data = {
    "Name" : ["Piyush", "Vai", "Arpit", "Rahul", "Sneha", "Rohan", "Priya"],
    "Age" : [20, 20, 19, 18, 19, 21, 18],
    "Gender" : ["Male", "Female", "Male", "Male", "Female", "Male", "Female"],
    "Marks" : [82, 88, 85, 60, 90, 78, 55],
    "Attendance" : [85.0, 95.0, 92.5, 70.0, 98.0, 88.0, 60.5],
    "Grade" : ["A", "A", "A", "C", "A", "B", "F"],
    "City" : ["Mumbai", "Pune", "Pune", "Mumbai", "Delhi", "Delhi", "Mumbai"]
}

df = pd.DataFrame(data)
print(df)
```

## 1. Line Plot
Used to show trends across students.

```python
sns.lineplot(data = df, x = 'Name', y = 'Marks', color = 'blue', marker = 'o')
plt.title("Student Marks - Line Plot")
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Two Lines

sns.lineplot(data=df, x="Name", y="Marks",      marker="o", label="Marks",      color="blue")
sns.lineplot(data=df, x="Name", y="Attendance",  marker="s", label="Attendance",  color="green")

plt.title("Marks vs Attendance - Line Plot")
plt.xlabel("Student Name")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## 2. Bar Plot
Used to compare values across categories.

```python
sns.barplot(data = df, x = 'Name' , y = 'Marks' , hue = 'Name', palette = 'rocket')
plt.xticks(rotation = 90)
plt.title("Student Marks - Bar Plot")
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.ylim(0, 110)
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

```python
sns.barplot(data=df, x="Name", y="Marks", hue="Gender", palette="Set1")
plt.xticks(rotation = 10)
plt.title("Marks by Gender - Bar Plot")
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.legend(title="Gender")
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

## 3. Histogram & Distribution Plot
Used to see how data is distributed.

```python
sns.histplot(data = df, x = 'City', bins = 5, hue='City', palette="Set2", kde = True)
plt.title("Students Distribution by Cities")
plt.xlabel("City")
plt.ylabel("Frequency")
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

```python
sns.kdeplot(data=df, x="Marks", fill=True, color="blue", alpha=0.5)
sns.kdeplot(data=df, x="Attendance", fill=True, color="green", alpha=0.5)

plt.title("Marks & Attendance - KDE Plot")
plt.xlabel("Value")
plt.grid(True)
plt.tight_layout()
plt.show()
```

## 4. Scatter Plot
Used to find relationship between two variables.

```python
sns.scatterplot(data=df, x="Attendance", y="Marks", hue="Gender", style="Gender", s=150, palette="Set1")
plt.title("Attendance vs Marks - Scatter Plot")
plt.xlabel("Attendance %")
plt.ylabel("Marks")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#scatter with names

sns.scatterplot(data=df, x="Attendance", y="Marks", s=150, color="purple")

for i, row in df.iterrows():
    plt.text(row["Attendance"] + 0.3, row["Marks"] + 0.5, row["Name"], fontsize=9)

plt.title("Attendance vs Marks with Names")
plt.xlabel("Attendance %")
plt.ylabel("Marks")
plt.grid(True)
plt.tight_layout()
plt.show()
```

## 5. Box Plot
Used to show spread, median and outliers.

```python
sns.boxplot(data=df[["Marks", "Attendance"]], palette="Set2")
plt.title("Marks & Attendance - Box Plot")
plt.ylabel("Value")
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

## 6. Violin Plot
Combination of box plot and KDE.
Shows distribution shape along with spread.

```python
sns.violinplot(data=df, x="Gender", y="Marks", palette="Set2")
plt.title("Marks by Gender - Violin Plot")
plt.xlabel("Gender")
plt.ylabel("Marks")
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

## 7. Count Plot
Used to count occurrences of categories.
Like a bar chart for categorical data.

```python
sns.countplot(data=df, x="City", palette="Set1")
plt.title("Students per City - Count Plot")
plt.xlabel("City")
plt.ylabel("Count")
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

```python
sns.countplot(data=df, x="City", hue="Gender", palette="Set1")

plt.title("Students per City by Gender")
plt.xlabel("City")
plt.ylabel("Count")
plt.legend(title="Gender")
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

## 8. Heatmap
Used to show correlation between numerical columns.
Darker color = stronger relationship.

```python
corr = df[["Age", "Marks", "Attendance"]].corr()

sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

## 9. Pair Plot
Shows relationships between all numerical columns at once.
Very useful in EDA.

```python
sns.pairplot(df[["Age", "Marks", "Attendance", "Gender"]], hue="Gender", palette="Set1")
plt.suptitle("Pair Plot - Student Data", y=1.02)
plt.tight_layout()
plt.show()
```

## 10. Strip Plot & Swarm Plot
Used to show individual data points by category.

```python
# Strip Plot
plt.figure(figsize=(8, 4))

sns.stripplot(data=df, x="City", y="Marks", hue="Gender",
              palette="Set1", size=10, jitter=True)

plt.title("Marks by City - Strip Plot")
plt.xlabel("City")
plt.ylabel("Marks")
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

```python
# Swarm Plot
plt.figure(figsize=(8, 4))

sns.swarmplot(data=df, x="City", y="Marks", hue="Gender",
              palette="Set2", size=10)

plt.title("Marks by City - Swarm Plot")
plt.xlabel("City")
plt.ylabel("Marks")
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

## 11. Seaborn Themes & Styles
Seaborn has 5 built in themes.
- darkgrid
- whitegrid
- dark
- white
- ticks

```python
styles = ["darkgrid", "whitegrid", "dark", "white", "ticks"]

plt.figure(figsize=(15, 8))

for i, style in enumerate(styles, 1):
    plt.subplot(2, 3, i)

    sns.set_style(style)
    plt.bar(df["Name"], df["Marks"])
    plt.title(style)
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
palettes = ["Set1", "Set2", "Set3", "coolwarm", "Blues", "husl"]

fig, axes = plt.subplots(2, 3, figsize=(14, 8))
axes = axes.flatten()

for i, palette in enumerate(palettes):
    colors = sns.color_palette(palette, len(df))
    axes[i].bar(df["Name"], df["Marks"], color=colors)
    axes[i].set_title(f"Palette: {palette}")
    axes[i].tick_params(axis="x", rotation=45)

plt.suptitle("Seaborn Color Palettes", fontsize=14)
plt.tight_layout()
plt.show()
```
