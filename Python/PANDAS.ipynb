# Pandas

Pandas is a Python library used for data manipulation and analysis.
It provides two main structures — Series and DataFrame.

- Works like Excel but in Python
- Handle missing data easily
- Filter, sort, group data with simple code
- Foundation for EDA and Machine Learning



```python
import pandas as pd
import numpy as np
print("Pandas version:", pd.__version__)
```

## Dataset
We are using a student dataset with 7 records.

Columns- Name, Age, Gender, Marks, Attendance, Grade, City

```python
data = {
    "Name" : ["Aarav", "Sneha", "Ravi", "Priya", "Arjun", "Meera", "Karan"],
    "Age" : [18, 19, 20, 17, 21, 19, 18],
    "Gender" : ["Male", "Female", "Male", "Female", "Male", "Female", "Male"],
    "Marks" : [72, 85, 60, 90, 55, 78, 88],
    "Attendance" : [85.0, 92.5, 70.0, 98.0, 60.5, 88.0, 95.0],
    "Grade" : ["B", "A", "C", "A", "F", "B", "A"],
    "City" : ["Mumbai", "Delhi", "Mumbai", "Pune", "Delhi", "Pune", "Mumbai"]
}
df = pd.DataFrame(data)
print(df)
```

## 1. DataFrame Properties

```python
print(df.head())
print(df.tail(5))
print(df.shape)
print(df.info())
print(df.describe())
print(df.columns)
print(df.dtypes)
```

## 2. Selecting Columns & Rows

```python
# single col
print(df['Name'])

# multiple cols
print(df[['Name', 'Marks', 'City']])

# select the rows by indexing - iloc
print(df.iloc[0])
print(df.iloc[0:3])
print(df.iloc[1 , 4])  # row 1 and column 4 means sneha's attendance

# select row by labels - loc
print(df.loc[0])
print(df.loc[0:3])
print(df.loc[0:4, 'Name':'Marks']) # rows 0 to 4 and columns from 'Name' to 'Marks'
```

## 3. Filtering Data
Filter students based on conditions.

```python
#students who have marks more than 70
passed = df[df['Marks'] > 70]
print(passed)

#students who have A grade
A_grade = df[df['Grade'] == 'A']
print(A_grade)

#Multiple conditions - student who have 70+ marks and 90+ attendance
good_students =  df[(df['Marks'] > 70) & (df['Attendance'] > 90)]
print(good_students)

#students from Mumbai
mumbai_students = df[df.City == 'Mumbai']
print(mumbai_students)

#isin() - students from Mumbai and Pune
M_P_students = df[df.City.isin(['Mumbai', 'Pune'])]
print(M_P_students)
```

## 4. Sorting Data

```python
#sort by Marks
print(df.sort_values('Marks'))

#descending
print(df.sort_values('Marks', ascending=False))

#multiple cols
print(df.sort_values(["Marks","Grade"], ascending = [False, True]))
```

## 5. Adding & Dropping Columns

```python
#add new column - total marks would be out of 125
df['Total'] = df.Marks + 25
print(df.head())

# Drop a column
df_dropped = df.drop(columns=["Total"])
print(df_dropped.columns)

# Drop a row
df_dropped_row = df.drop(index=0)
print(df_dropped_row)

# Add Pass/Fail column
df["Result"] = df["Marks"].apply(lambda x: "Pass" if x >= 70 else "Fail")
print(df[["Name", "Marks", "Result"]])
```

## 6. Handling Missing Values

```python
# Making some missing values
df_missing = df.copy()
df_missing.loc[1, "Marks"] = None
df_missing.loc[3, "Attendance"] = None
df_missing.loc[5, "Grade"] = None

print(df_missing)

# Check missing values
print(df_missing.isnull())
print(df_missing.isnull().sum())

# Fill missing marks with mean
df_missing["Marks"] = df_missing["Marks"].fillna(df_missing["Marks"].mean())

# Fill missing attendance with 0
df_missing["Attendance"] = df_missing["Attendance"].fillna(0)

# Drop rows where Grade is missing
df_clean = df_missing.dropna(subset=["Grade"])

print(df_clean)
```

## 7. Statistical Operations

```python
print("Mean Marks       :", df["Marks"].mean())
print("Median Marks     :", df["Marks"].median())
print("Max Marks        :", df["Marks"].max())
print("Min Marks        :", df["Marks"].min())
print("Std Deviation    :", df["Marks"].std())
print("Total Marks Sum  :", df["Marks"].sum())
print("Value Counts     :\n", df["Grade"].value_counts())
print("Unique Cities    :", df["City"].unique())
print("City Count       :", df["City"].nunique())
```

## 8. GroupBy
Group data and find insights.

```python
# Avg Marks by Gender
print(df.groupby('Gender')['Marks'].mean())

# Avg Marks by City
print("\n",df.groupby('City')['Marks'].mean())

# students in each city
print("\n", df.groupby('City')['Name'].count())

# Multiple aggreation
print(df.groupby('City').agg(
    Avg_marks = ('Marks', 'mean'),
    Min_marks = ('Marks', 'min'),
    Max_marks = ('Marks', 'max'),
    Total_students = ('Name', 'count')
))
```

## 9. String Operations
Working with text columns.

```python
# Uppercase names
print(df["Name"].str.upper())

# Lowercase cities
print(df["City"].str.lower())

# Name starts with A
print(df[df["Name"].str.startswith("A")])

# Name contains 'a' (case insensitive)
print(df[df["Name"].str.contains("a", case=False)])

# Length of each name
print(df["Name"].str.len())
```

## 10. Apply & Map
Apply custom logic to columns.

```python
# apply() → custom grade logic
def assign_grade(mark):
    if mark >= 85:
        return "Distinction"
    elif mark >= 70:
        return "Pass"
    else:
        return "Fail"

df["New_Grade"] = df["Marks"].apply(assign_grade)
print(df[["Name", "Marks", "New_Grade"]])

# map() → map gender to numbers
df["Gender_Code"] = df["Gender"].map({"Male": 0, "Female": 1})
print(df[["Name", "Gender", "Gender_Code"]])
```

## 11. Renaming & Resetting Index

```python
# Rename columns
df_renamed = df.rename(columns={"Marks": "Score", "Attendance": "Present_%"})
print(df_renamed.columns)

# Reset index
df_reset = df.reset_index(drop=True)
print(df_reset)

# Set Name as index
df_indexed = df.set_index("Name")
print(df_indexed)
```

## 12. Merging & Concatenating DataFrames

```python
# Two separate dataframes
df1 = df[["Name", "Marks"]].iloc[:4]
df2 = df[["Name", "Marks"]].iloc[4:]

# Concatenate → join vertically
combined = pd.concat([df1, df2], ignore_index=True)
print(combined)

# Merge → like SQL join
extra_info = pd.DataFrame({
    "Name"   : ["Aarav", "Sneha", "Ravi", "Priya", "Arjun", "Meera", "Karan"],
    "Subject": ["Maths", "Science", "Maths", "English", "Science", "Maths", "English"]
})

merged = pd.merge(df, extra_info, on="Name")
print(merged[["Name", "Marks", "Subject"]])
```

## 13. Save & Load Data

```python
# Save to CSV
df.to_csv("students.csv", index=False)
print("Saved!")

# Load from CSV
df_loaded = pd.read_csv("students.csv")
print(df_loaded.head())
```
