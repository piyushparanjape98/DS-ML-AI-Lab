# Matplotlib

Matplotlib is a Python library used for data visualization.
It helps us create charts and graphs from data.

- Visualize data quickly
- Supports many chart types
- Highly customizable
- Works well with NumPy and Pandas

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Dataset
Same student dataset used throughout this notebook.

```python
data = {
    "Name" : ["Piyush", "Vai", "Arpit", "Rahul", "Sneha", "Rohan", "Priya"],
    "Age" : [20, 20, 19, 18, 19, 21, 18],
    "Gender" : ["Male", "Female", "Male", "Male", "Female", "Male", "Female"],
    "Marks" : [82, 88, 85, 60, 90, 78, 55],
    "Attendance" : [85.0, 95.0, 92.5, 70.0, 98.0, 88.0, 60.5],
    "Grade" : ["A", "A", "A", "C", "A", "B", "F"],
    "City" : ["Mumbai", "Pune", "Pune", "Mumbai", "Delhi", "Delhi", "Mumbai"]
}

df = pd.DataFrame(data)
print(df)
```

## Line Plot
Used to show trends over a sequence.

```python
plt.plot(df.Name , df.Marks, color = 'Blue', marker = 'o', linewidth = 2, markersize = 8)
plt.title("Student Marks")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.grid()
plt.show()
```

## Two Line plot

```python
from matplotlib.lines import lineStyles
plt.plot(df['Name'], df['Marks'], color = 'Blue', marker = '*' , linewidth = 2 ,)
plt.plot(df.Name, df.Attendance, color = 'Orange', marker = '+' , linestyle = '--')
plt.title('Marks vs Attendance')
plt.xlabel('Name')
plt.ylabel('Marks and Attendance')
plt.grid()
plt.show()
```

## 2. Bar Chart
Used to compare values.

```python
plt.bar(df['Name'], df['Marks'], color = ['Red', 'Orange', 'Yellow'], edgecolor = 'Black', )
plt.title('Student Marks')
plt.ylabel('Marks')
plt.xlabel('Name')
plt.ylim(0,100)
plt.grid(axis = 'y')
plt.tight_layout()
plt.show()
```

```python
#Horizontal bar chart

plt.barh(df['Name'], df['Marks'], color = ['Red', 'Orange', 'Yellow'], edgecolor = 'Black', )
plt.title('Student Marks')
plt.ylabel('Name')
plt.xlabel('Marks')
plt.grid(axis = "x")
plt.tight_layout()
plt.show()
```

```python
#Grouped Bar Chart

x = np.arange(len(df["Name"]))
width = 0.35

plt.bar(x - width/2, df["Marks"],      width, label="Marks",      color="steelblue", edgecolor="black")
plt.bar(x + width/2, df["Attendance"], width, label="Attendance",  color="coral",     edgecolor="black")
plt.title("Marks vs Attendance - Grouped Bar Chart")
plt.xlabel("Student Name")
plt.ylabel("Value")
plt.xticks(x, df["Name"])
plt.legend()
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

## 3. Histogram
Used to show distribution of data.

```python
plt.hist(df['City'], bins = 5, color='steelblue', edgecolor = 'black',)
plt.title('Student Cities')
plt.xlabel('City')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

## 4. Scatter Plot
Used to find relationship between two variables.

```python
plt.scatter(df.Attendance, df.Marks, color='purple',  s=100)

# Annotate each point with student name
for i, name in enumerate(df["Name"]):
    plt.annotate(name, (df["Attendance"][i], df["Marks"][i]),
                 textcoords="offset points", xytext=(5, 5), fontsize=9)

plt.title('Student Marks vs Attendance')
plt.xlabel('Attendance')
plt.ylabel('Marks')
plt.tight_layout()
plt.show()
```

## 5. Pie Chart
Used to show proportions.

```python
grade_counts = df["Grade"].value_counts()

plt.figure(figsize=(6, 6))

plt.pie(grade_counts.values, labels=grade_counts.index, autopct="%1.1f%%", colors=["steelblue", "coral", "green", "orange"], startangle=90,
        explode=[0.05] * len(grade_counts))

plt.title("Grade Distribution - Pie Chart")
plt.tight_layout()
plt.show()
```

## 6. Box Plot
Used to show spread and outliers in data.

```python
plt.figure(figsize=(8, 4))

plt.boxplot([df["Marks"], df["Attendance"]], labels=["Marks", "Attendance"], patch_artist=True,
            boxprops=dict(facecolor="steelblue", color="black"),
            medianprops=dict(color="lightGreen", linewidth=2))

plt.title("Marks & Attendance - Box Plot")
plt.ylabel("Value")
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

## 7. Subplots
Display multiple charts in one figure.

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Line Plot
axes[0, 0].plot(df["Name"], df["Marks"], color="blue", marker="o")
axes[0, 0].set_title("Marks - Line Plot")
axes[0, 0].grid(True)

# Bar Chart
axes[0, 1].bar(df["Name"], df["Marks"], color="steelblue", edgecolor="black")
axes[0, 1].set_title("Marks - Bar Chart")
axes[0, 1].grid(axis="y")

# Scatter Plot
axes[1, 0].scatter(df["Attendance"], df["Marks"], color="purple", s=100, edgecolor="black")
axes[1, 0].set_title("Attendance vs Marks - Scatter")
axes[1, 0].grid(True)

# Pie Chart
grade_counts = df["Grade"].value_counts()
axes[1, 1].pie(grade_counts.values, labels=grade_counts.index,
               autopct="%1.1f%%", startangle=90)
axes[1, 1].set_title("Grade Distribution - Pie")

plt.suptitle("Student Data - All Charts", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()
```

## 8. Styling & Customization
Adding colors, styles and themes to charts.

```python
# Available styles
print(plt.style.available)
```

```python
plt.style.use("ggplot")

plt.bar(df["Name"], df["Marks"], color="steelblue", edgecolor="black")

plt.title("Marks - GGPlot Style", fontsize=14, fontweight="bold")
plt.xlabel("Student Name", fontsize=11)
plt.ylabel("Marks", fontsize=11)
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

Adding Annotations

```python
plt.style.use("default")

bars = plt.bar(df["Name"], df["Marks"], color="steelblue", edgecolor="black")

# Add value on top of each bar
for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2,
             bar.get_height() + 1,
             str(int(bar.get_height())),
             ha="center", va="bottom", fontsize=10, fontweight="bold")

plt.title("Marks with Values on Bars")
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.ylim(0, 100)
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

## 9. Saving Charts
Save charts as image files.

```python
plt.bar(df["Name"], df["Marks"], color="steelblue", edgecolor="black")

plt.title("Student Marks")
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.grid(axis="y")
plt.tight_layout()

# Save as PNG
plt.savefig("student_marks.png", dpi=150)
plt.show()
print("Chart saved!")
```
