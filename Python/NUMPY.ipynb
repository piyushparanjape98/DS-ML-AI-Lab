# NumPy

**What is NumPy?**

NumPy (Numerical Python) is a library used for numerical computations.
It provides support for arrays, matrices, and many mathematical functions.

**Why NumPy?**
- Faster than Python lists
- Less memory usage
- Built-in mathematical operations
- Foundation for Pandas, Scikit-learn, etc.

```python
import numpy as np
print("NumPy version:", np.__version__)
```

## Dataset
We will create and use this small student dataset which contain **ages** , **marks (out of 100)** , **attendance (in percentage)** throughout this notebook for the numpy operations.


```python
ages = np.array([18, 19, 20, 17, 21, 19, 18])
marks = np.array([72, 85, 60, 90, 55, 78, 88])
attendance = np.array([85.0, 92.5, 70.0, 98.0, 60.5, 88.0, 95.0])

names = np.array(["Piyush", "Vai", "Arpit", "Om", "Kabir", "Naina", "Parth"])

print("Names     :", names)
print("Ages      :", ages)
print("Marks     :", marks)
print("Attendance:", attendance)
```

## 1. Array Properties
Understanding shape, size, dimensions and datatype of an array.

```python
print("Array Properties")

print("\nMarks array:", marks)
print("Shape      :", marks.shape)       # (7,) → 7 elements, 1D
print("Size       :", marks.size)        # total elements = 7
print("Dimensions :", marks.ndim)        # 1D array
print("Data Type  :", marks.dtype)       # int64
print("Item Size  :", marks.itemsize, "bytes")  # bytes per element
```

## 2. Creating Arrays — Different Methods

```python
# Zeros
empty_scores = np.zeros(7)
print("Zero scores:", empty_scores)

# Ones
passed = np.ones(7, dtype=int)
print("Passed     :", passed)

# Full
default_marks = np.full(7, 50)
print("Default    :", default_marks)

# arange
roll_numbers = np.arange(1, 8)
print("Roll Nos   :", roll_numbers)

# linspace
grade_scale = np.linspace(0, 100, 5)
print("Grade Scale:", grade_scale)

# Random integers
np.random.seed(42)
quiz_marks = np.random.randint(1, 11, size=7)
print("Quiz Marks :", quiz_marks)
```

## 3. Indexing & Slicing
Accessing specific elements from the array.

```python
print("All marks     :", marks)
print("First student :", marks[0])
print("Last student  :", marks[-1])      # Karan
print("First 3 marks :", marks[:3])      # Aarav, Sneha, Ravi
print("Last 3 marks  :", marks[-3:])     # Arjun, Meera, Karan
print("Alternate     :", marks[::2])

# Fancy Indexing → pick specific students
selected = marks[[0, 2, 5]]
print("Piyush, Arpit, Naina marks:", selected)
```

## 4. Basic Math Operations
Arithmetic operations applied on arrays element-wise.

```python
print("Original Marks:", marks)

# Add 5 bonus marks to everyone
bonus = marks + 5
print("After Bonus (+5) :", bonus)

# Deduct 2 for late submission
penalty = marks - 2
print("After Penalty (-2):", penalty)

# Double the marks (for weightage)
doubled = marks * 2
print("Doubled marks    :", doubled)

# Percentage out of 200 total
percentage = marks / 200 * 100
print("Percentage       :", percentage)

# Marks squared
squared = marks ** 2
print("Squared          :", squared)
```

```python
# Internal marks + External marks = Total
internal = np.array([20, 25, 18, 28, 15, 22, 26])
external = np.array([72, 85, 60, 90, 55, 78, 88])

total = internal + external
print("Internal:", internal)
print("External:", external)
print("Total   :", total)

difference = external - internal
print("Difference:", difference)
```

## 5. Statistical Operations
Descriptive statistics on marks and attendance.

```python
print("Marks Statistics")
print("Marks         :", marks)
print("Mean          :", np.mean(marks))
print("Median        :", np.median(marks))
print("Mode-like Max :", np.max(marks))
print("Minimum       :", np.min(marks))
print("Maximum       :", np.max(marks))
print("Sum           :", np.sum(marks))
print("Std Deviation :", np.std(marks).round(2))
print("Variance      :", np.var(marks).round(2))
print("Range         :", np.max(marks) - np.min(marks))

print("\nAttendance Statistics")
print("Attendance    :", attendance)
print("Average Attend:", np.mean(attendance).round(2))
print("Lowest Attend :", np.min(attendance))
print("Highest Attend:", np.max(attendance))
```

```python
print("25th Percentile (Q1):", np.percentile(marks, 25))
print("50th Percentile (Q2):", np.percentile(marks, 50))
print("75th Percentile (Q3):", np.percentile(marks, 75))
print("Top 10% threshold   :", np.percentile(marks, 90))
```

## 6. Filtering — Boolean Masking
Filter students based on conditions.

```python
print("Marks:", marks)
print("Names:", names)

# Students who passed (marks >= 70)
passed_mask = marks >= 70
print("\nPassed mask   :", passed_mask)
print("Passed students:", names[passed_mask])
print("Their marks   :", marks[passed_mask])

# Students who failed
failed_mask = marks < 70
print("\nFailed students:", names[failed_mask])
print("Their marks    :", marks[failed_mask])

# Students with attendance below 75
low_attendance = attendance < 75
print("\nLow Attendance:", names[low_attendance])

# Multiple conditions → failed AND low attendance
at_risk = (marks < 70) & (attendance < 75)
print("\nAt Risk Students:", names[at_risk])
```

## 7. Sorting
Sort marks and find rank order.

```python
print("Original Marks:", marks)

# Sort marks ascending
sorted_marks = np.sort(marks)
print("Sorted (Low→High):", sorted_marks)

# Sort descending
sorted_desc = np.sort(marks)[::-1]
print("Sorted (High→Low):", sorted_desc)

# argsort → returns index positions in sorted order
rank_order = np.argsort(marks)[::-1]
print("\nRank Order (indices):", rank_order)
print("Topper to Lowest:")
for rank, i in enumerate(rank_order):
    print(f"  Rank {rank+1}: {names[i]} → {marks[i]}")
```

## 8. Reshaping Arrays
Converting 1D arrays into 2D matrix format.

```python
# Combine marks of 3 subjects for 5 students → 15 values
all_marks = np.array([72, 85, 60, 90, 55,   # Subject 1
                      78, 88, 70, 65, 80,   # Subject 2
                      88, 76, 55, 92, 68])  # Subject 3

print("Original (1D):", all_marks)
print("Shape         :", all_marks.shape)

# Reshape to 3 subjects × 5 students
matrix = all_marks.reshape(3, 5)
print("\nReshaped (3x5):\n", matrix)
print("Shape:", matrix.shape)

# Transpose → 5 students × 3 subjects
transposed = matrix.T
print("\nTransposed (5x3):\n", transposed)
```

```python
print("Marks Matrix (3 subjects x 5 students):\n", matrix)

# Average marks per student (column-wise)
avg_per_student = np.mean(matrix, axis=0)
print("\nAvg per student:", avg_per_student.round(2))

# Average marks per subject (row-wise)
avg_per_subject = np.mean(matrix, axis=1)
print("Avg per subject:", avg_per_subject.round(2))

# Total marks per student
total_per_student = np.sum(matrix, axis=0)
print("Total per student:", total_per_student)
```

## 9. Useful NumPy Functions

```python
# where() → assign grades
grades = np.where(marks >= 75, "Distinction", "Pass")
print("Grades:", grades)

# clip() → cap marks between 60 and 90
clipped = np.clip(marks, 60, 90)
print("\nOriginal Marks :", marks)
print("Clipped (60-90):", clipped)

# unique() → unique ages
print("\nAll Ages   :", ages)
print("Unique Ages:", np.unique(ages))

# cumsum() → running total of marks
cumulative = np.cumsum(marks)
print("\nCumulative Marks:", cumulative)

# diff() → difference between consecutive marks
differences = np.diff(marks)
print("Differences     :", differences)
```

## 10. Math Functions

```python
print("Marks:", marks)

# Square root of marks
print("Square Root    :", np.sqrt(marks).round(2))

# Log of marks (natural log)
print("Log (natural)  :", np.log(marks).round(2))

# Log base 10
print("Log base 10    :", np.log10(marks).round(2))

# Absolute values
changes = np.array([-5, 3, -8, 10, -2, 7, -1])
print("\nChanges        :", changes)
print("Absolute Values:", np.abs(changes))

# Rounding attendance
print("\nAttendance (raw)    :", attendance)
print("Rounded (0 decimal) :", np.round(attendance, 0))
print("Floor               :", np.floor(attendance))
print("Ceil                :", np.ceil(attendance))
```
