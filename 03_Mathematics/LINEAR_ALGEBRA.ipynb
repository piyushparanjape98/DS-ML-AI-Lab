## Property 1 : Transpose of a Transposed Matrix is the Original Matrix

```python
m1
```

```python
transpose = m1.T
transpose
```

```python
transpose.T
```

```python
if np.array_equal(m1, transpose.T):
  print("Property 1 is satisfied")
else:
  print("Property 1 is not satisfied")
```

## Property 2: (Additive property) : (A+B).T = A.T + B.T

```python
print(a)
print(b)
```

```python
c = a + b
lhs = c.T
lhs
```

```python
d = a.T + b.T
rhs = d
rhs
```

```python
if np.array_equal(lhs, rhs):
  print("Property 2 is satisfied")
else:
  print("Property 2 is not satisfied")
```

## Property 3: (Multiplicative Property): (A * B).T = B.T * A.T

```python
print(a)
print(b)
```

```python
c = a * b
lhs = c.T
lhs
```

```python
d = b.T * a.T
a
```

```python
if (lhs == d).all:
  print("Property 3 is satisfied")
else:
  print("Property 3 is not satisfied")


# .all means - all the elements of both matrices would be compared
# .any means - if any one of the element of bith matrices is same or not
```

another method of property 3

```python
p = np.random.randint(10, size = (2,2))
q = np.random.randint(10, size = (2,2))
if np.array_equal((p@q).T , (q.T)@(p.T)):
  print("hence proved")
else:
  print("not proved")
```

## The Dot Product of the Original Matriz and its Inverse Matrix is equal to the IDENTITY MATRIX

```python
m1
```

```python
m1_inverse = np.linalg.inv(m1)
m1_inverse
```

```python
res = m1 @ m1_inverse
res
```

```python
np.eye(3)
```

```python
if np.allclose(res, np.eye(3)):
  print("our dot product assumption is True")
else:
  print("our dot product assumption is False")


  # .allclose --> check the two arrays/Matrices are element wise equal or not
  # (np.allclose is used instead of == np.array_equal when comapring floating point number)
```

## Solving System of Linear Equations (Ax = b)

```python
A = np.array([[2, 1, 1],
              [1, 3, 2],
              [1, 0, 0]])

b = np.array([4, 5, 6])

x = np.linalg.solve(A, b)
x
```

```python
# Verify
np.allclose(A @ x, b)
```
