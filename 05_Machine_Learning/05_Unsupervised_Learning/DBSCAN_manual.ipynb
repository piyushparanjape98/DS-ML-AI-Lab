DBSCAN (Manual Implementation)

```python
import numpy as np

def euclidean_distance(a, b):
    return np.sqrt(np.sum((a - b) ** 2))

def get_neighbors(X, point_idx, eps):
    neighbors = []
    for i in range(len(X)):
        if euclidean_distance(X[point_idx], X[i]) <= eps:
            neighbors.append(i)
    return neighbors

def dbscan(X, eps, min_samples):
    labels = np.full(len(X), -1)  # -1 = noise, by default
    visited = np.zeros(len(X), dtype=bool)
    cluster_id = 0

    for i in range(len(X)):
        if visited[i]:
            continue
        visited[i] = True
        neighbors = get_neighbors(X, i, eps)

        if len(neighbors) < min_samples:
            continue  # stays noise for now

        # start a new cluster
        labels[i] = cluster_id
        seeds = neighbors.copy()

        while seeds:
            current = seeds.pop()
            if not visited[current]:
                visited[current] = True
                current_neighbors = get_neighbors(X, current, eps)
                if len(current_neighbors) >= min_samples:
                    seeds.extend(current_neighbors)  # expand cluster further
            if labels[current] == -1:
                labels[current] = cluster_id

        cluster_id += 1

    return labels

# Example usage
X_demo = np.array([[1, 1], [1.2, 1.1], [1, 1.3], [8, 8]])
labels = dbscan(X_demo, eps=0.5, min_samples=2)
print("Cluster labels:", labels)
```
