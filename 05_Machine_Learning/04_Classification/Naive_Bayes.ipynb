# Naive Bayes (Text Classification)

- Naive Bayes: predicts class probability using Bayes' Theorem, assuming features (words) are independent
- Used here as Multinomial Naive Bayes â built for word counts/frequencies, not continuous numbers
- Dataset: spam.csv (predicting spam vs ham SMS messages)[kaggle.com/code/karnikakapoor/spam-or-ham-sms-classifier/input](https://)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/spam.csv', encoding='latin-1')
df.head()
```

```python
df = df[["v1", "v2"]]
df.columns = ["label", "message"]

# Encode label: ham = 0, spam = 1
df["label"] = df["label"].map({"ham": 0, "spam": 1})

print(df.head())
print(df["label"].value_counts())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix, accuracy_score
import warnings
warnings.filterwarnings("ignore")
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    df["message"], df["label"], test_size=0.2, random_state=42
)
print("Train size:", X_train.shape[0])
print("Test size :", X_test.shape[0])

cv = CountVectorizer()
X_train_vec = cv.fit_transform(X_train)
X_test_vec = cv.transform(X_test)
print("Vocabulary size:", len(cv.vocabulary_))

model = MultinomialNB()
model.fit(X_train_vec, Y_train)
Y_pred = model.predict(X_test_vec)
print(pd.DataFrame({"Actual": Y_test.values, "Predicted": Y_pred}))

cm = confusion_matrix(Y_test, Y_pred)
acc = accuracy_score(Y_test, Y_pred)
print("Confusion Matrix:\n", cm)
print("Accuracy:", round(acc, 4))
```

```python
custom_message = ["Congratulations! You won a free prize, claim now!"]
custom_vec = cv.transform(custom_message)
print("Prediction (1=spam, 0=ham):", model.predict(custom_vec)[0])
```

## Conclusion
- Each SMS was converted into word counts (CountVectorizer) before training, since Naive Bayes needs numeric input
- Multinomial Naive Bayes calculated, for each message, which class (spam/ham) was more probable given its word counts
- Confusion Matrix & Accuracy show how well it caught spam on unseen messages
- Tried a custom message at the end to see the model classify new, unseen text
