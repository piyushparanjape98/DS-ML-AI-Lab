SVM (Manual Implementation)

```python
import numpy as np

def train_svm(X, y, lr=0.001, lambda_param=0.01, epochs=1000):
    n_samples, n_features = X.shape
    y_ = np.where(y <= 0, -1, 1)  # SVM math expects labels as -1/+1, not 0/1

    w = np.zeros(n_features)
    b = 0

    for epoch in range(epochs):
        for i in range(n_samples):
            condition = y_[i] * (np.dot(X[i], w) + b) >= 1
            if condition:
                # point correctly classified outside margin -> only regularization term
                dw = 2 * lambda_param * w
                db = 0
            else:
                # point inside margin or misclassified -> hinge loss term added
                dw = 2 * lambda_param * w - y_[i] * X[i]
                db = -y_[i]
            w -= lr * dw
            b -= lr * db
    return w, b

def predict_svm(X, w, b):
    return np.sign(np.dot(X, w) + b)

# Example usage
X_demo = np.array([[1, 1], [2, 1], [1.5, 0.5], [5, 5], [6, 5], [5.5, 4.5]])
y_demo = np.array([0, 0, 0, 1, 1, 1])  # 2 separable clusters

w, b = train_svm(X_demo, y_demo)
preds = predict_svm(X_demo, w, b)
print("Weights:", w, "Bias:", b)
print("Predictions:", np.where(preds == -1, 0, 1))
```
