# Support Vector Machine (SVM)

- SVM: finds the hyperplane that separates classes with the maximum margin
- Uses kernel trick (e.g. RBF) to handle data that isn't linearly separable
- Dataset: breast_cancer.csv (predicting diagnosis — malignant vs benign)
[https://www.kaggle.com/code/mragpavank/breast-cancer-wisconsin/input](https://)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/data.csv')
df
```

```python
# Drop ID column (no predictive value) and any empty trailing column
df = df.drop(columns=[col for col in ["id", "Unnamed: 32"] if col in df.columns])

# Encode diagnosis: M = 1 (malignant), B = 0 (benign)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["diagnosis"] = le.fit_transform(df["diagnosis"])

X = df.drop("diagnosis", axis=1)
Y = df["diagnosis"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, accuracy_score
import warnings
warnings.filterwarnings("ignore")
```

(Scaling matters a lot for SVM — it relies on distances/margins, so features on different scales would distort the boundary.)

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
print("Train size:", X_train.shape[0])
print("Test size :", X_test.shape[0])

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

model = SVC(kernel="rbf", random_state=42)
model.fit(X_train, Y_train)

Y_pred = model.predict(X_test)
print(pd.DataFrame({"Actual": Y_test.values, "Predicted": Y_pred}))

cm = confusion_matrix(Y_test, Y_pred)
acc = accuracy_score(Y_test, Y_pred)
print("Confusion Matrix:\n", cm)
print("Accuracy:", round(acc, 4))
```

## Conclusion
- SVM (with RBF kernel) found the boundary that best separated malignant vs benign cases with maximum margin
- Feature scaling was essential since the 30 features are on very different scales (e.g. area vs smoothness)
- Confusion Matrix & Accuracy show how well it predicted on unseen tumor samples
- RBF kernel handled non-linear separation better than a plain linear boundary would
