## Naive Bayes (Manual Implementation)

```python
import numpy as np

def gaussian_likelihood(x, mean, var):
    eps = 1e-6  # avoid division by zero
    coeff = 1 / np.sqrt(2 * np.pi * var + eps)
    exponent = np.exp(-((x - mean) ** 2) / (2 * var + eps))
    return coeff * exponent

def fit_naive_bayes(X_train, Y_train):
    classes = np.unique(Y_train)
    stats = {}
    for c in classes:
        X_c = X_train[Y_train == c]
        stats[c] = {
            "mean": X_c.mean(axis=0),
            "var": X_c.var(axis=0),
            "prior": X_c.shape[0] / X_train.shape[0]
        }
    return stats

def predict_naive_bayes(stats, x_query):
    posteriors = {}
    for c, params in stats.items():
        likelihood = np.prod(gaussian_likelihood(x_query, params["mean"], params["var"]))
        posteriors[c] = likelihood * params["prior"]
    return max(posteriors, key=posteriors.get)

# Example usage
X_train_demo = np.array([[25], [30], [45], [50], [55]])
Y_train_demo = np.array([0, 0, 1, 1, 1])

stats = fit_naive_bayes(X_train_demo, Y_train_demo)
prediction = predict_naive_bayes(stats, np.array([48]))
print("Predicted class:", prediction)
```
