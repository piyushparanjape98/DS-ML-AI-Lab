# Simple Linear Regression - Complete Practice Notebook

**What is Simple Linear Regression?**

Simple Linear Regression models the relationship between **one independent
variable (X)** and **one dependent variable (Y)** by fitting a straight line
through the data.

**Equation of the line:**

Y = mX + c

- Y → Predicted value (dependent variable)
- X → Input value (independent variable)
- m → Slope (how much Y changes for a unit change in X)
- c → Intercept (value of Y when X = 0)

**Goal:**
Find the best-fit line that minimizes the distance (error) between the
actual data points and the predicted line. This is done using the
**Least Squares Method**.

**Where Correlation ends and Regression begins:**
Correlation (covered earlier) tells us *if* and *how strongly* two variables
are related. Regression goes a step further — it builds a model that lets
us *predict* Y from X.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
import warnings
warnings.filterwarnings("ignore")
print("Libraries imported successfully")
```

## 1. Theory - Least Squares Method

**Formulas:**

m (slope)     = Σ[(x - x̄)(y - ȳ)] / Σ[(x - x̄)²]

c (intercept) = ȳ - m * x̄

These formulas find the line that minimizes the **sum of squared errors**
(the vertical distance between each actual point and the predicted line).

**Why "squared"?**
Squaring removes negative signs (so errors above and below the line don't
cancel out) and penalizes larger errors more heavily.

## 2. Quick Look at the Relationship

Before fitting any model, always visualize the raw data first.

```python
plt.figure(figsize=(8, 4))
plt.scatter(df["YearsExperience"], df["Salary"], color="steelblue", edgecolor="black", s=70)
plt.title("Years of Experience vs Salary (Raw Data)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.grid(True)
plt.tight_layout()
plt.show()
```

## 3. Manual Calculation (Least Squares - For Understanding)

This part is genuinely useful to do by hand once — it shows exactly where
sklearn's slope and intercept numbers come from.

```python
x = df["YearsExperience"]
y = df["Salary"]

x_mean = x.mean()
y_mean = y.mean()

numerator   = sum((x - x_mean) * (y - y_mean))
denominator = sum((x - x_mean)**2)

m_manual = numerator / denominator
c_manual = y_mean - (m_manual * x_mean)

print(f"Mean of X (Years Experience) : {x_mean:.4f}")
print(f"Mean of Y (Salary)           : {y_mean:.4f}")
print(f"\nManual Slope (m)     : {m_manual:.4f}")
print(f"Manual Intercept (c) : {c_manual:.4f}")
print(f"\nManual Regression Equation: Salary = {m_manual:.2f} * Experience + {c_manual:.2f}")
```

## 4. Fitting with Scikit-learn

Now let's confirm the manual result using sklearn's `LinearRegression`.

```python
X = df[["YearsExperience"]]   # sklearn expects X as 2D
Y = df["Salary"]

model = LinearRegression()
model.fit(X, Y)

m_sklearn = model.coef_[0]
c_sklearn = model.intercept_

print(f"Sklearn Slope (m)     : {m_sklearn:.4f}")
print(f"Sklearn Intercept (c) : {c_sklearn:.4f}")
print(f"\nSklearn Regression Equation: Salary = {m_sklearn:.2f} * Experience + {c_sklearn:.2f}")

print(f"\n=== Comparison ===")
print(f"Manual  -> m = {m_manual:.4f}, c = {c_manual:.4f}")
print(f"Sklearn -> m = {m_sklearn:.4f}, c = {c_sklearn:.4f}")
print("\nBoth match — confirms the least squares formula and sklearn use the same math.")
```

## 5. Making Predictions

Use the fitted model to predict salary for given years of experience.

```python
df["Predicted_Salary"] = model.predict(X)
print(df[["YearsExperience", "Salary", "Predicted_Salary"]].head(10))

# Predict for a new, unseen value
new_experience = [[5]]
predicted_salary = model.predict(new_experience)
print(f"\nPredicted Salary for 5 years of experience: {predicted_salary[0]:.2f}")
```

## 6. Visualization - Best Fit Line

```python
plt.figure(figsize=(8, 4))

plt.scatter(df["YearsExperience"], df["Salary"], color="steelblue",
            edgecolor="black", s=70, label="Actual Data")
plt.plot(df["YearsExperience"], df["Predicted_Salary"], color="red",
          linewidth=2, label="Best Fit Line")

plt.title(f"Years of Experience vs Salary\nSalary = {m_sklearn:.2f} * Experience + {c_sklearn:.2f}")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
