## Heirarchical Clustering (Agglomerative_Clustering) -  Manual Implementation

```python
import numpy as np

def euclidean_distance(a, b):
    return np.sqrt(np.sum((a - b) ** 2))

def cluster_distance_single_linkage(cluster_a, cluster_b):
    return min(euclidean_distance(a, b) for a in cluster_a for b in cluster_b)

def agglomerative_clustering(X, n_clusters):
    # start: every point is its own cluster
    clusters = [[point] for point in X]

    while len(clusters) > n_clusters:
        min_dist = float("inf")
        merge_i, merge_j = -1, -1

        # find the closest pair of clusters
        for i in range(len(clusters)):
            for j in range(i + 1, len(clusters)):
                dist = cluster_distance_single_linkage(clusters[i], clusters[j])
                if dist < min_dist:
                    min_dist = dist
                    merge_i, merge_j = i, j

        # merge the closest pair
        clusters[merge_i].extend(clusters[merge_j])
        del clusters[merge_j]

    return clusters

# Example usage
X_demo = np.array([[1, 1], [1.5, 2], [8, 8], [9, 9]])
result = agglomerative_clustering(X_demo, n_clusters=2)

for idx, c in enumerate(result):
    print(f"Cluster {idx}: {c}")
```
