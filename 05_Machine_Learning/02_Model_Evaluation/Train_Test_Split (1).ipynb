# Train-Test Split

- Train_Test_Split: train model on one part, test on unseen part
- Avoids overfitting (model memorizing instead of learning)


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("/content/50_Startups[1].csv")
df_encoded = pd.get_dummies(df, columns=["State"], drop_first=True)

X = df_encoded.drop("Profit", axis=1)
Y = df_encoded["Profit"]
df_encoded.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

print("Train size:", X_train.shape[0])
print("Test size :", X_test.shape[0])
```

```python
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)

print(pd.DataFrame({"Actual": Y_test, "Predicted": Y_pred}))
```

## Conclusion
- Split data → train → predict on unseen test set
- This separates training data from testing data so we can check
  how well the model performs on data it has never seen before
